# Rename coulmns name

```python
dataset.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
dataset = dataset[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
dataset.head()
```

```python
df=dataset
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
EE=pd.DataFrame(df['Year'].value_counts())
EE
```

```python
df["Wound"]=df["Wound"].fillna(0)
df["kill"]=df["kill"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df["casualities"]=df["kill"]+df["Wound"]
```

# Number of Total Attacks in Each Years

```python
df["Year"].value_counts()
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(df['Year'].value_counts().index, df['Year'].value_counts().values)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

```python
year = df['Year'].unique()
years_count = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=df,palette=None,edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

# Total Terrorist Attacks in Region

```python
plt.figure(figsize=(23,16),dpi=100)
ax=sns.countplot(x='Region',data=df)
plt.xlabel=("Total Terrorist Activities in Region")
plt.ylabel=("Total Number of attckes")
for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(df['Region'].value_counts().index, df['Region'].value_counts().values)
plt.xticks(rotation=90)
plt.show()
```

# Attacking Methods by Terrorists

```python
plt.subplots(figsize=(15,6))
sns.countplot('Attacktype',data=df, palette=None,order=df['Attacktype'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Attacking Methods by Terrorists')
plt.show()
```

# Terrorist Attack type

```python
df["Attacktype"].value_counts()
```

```python
plt.figure(figsize=(15,12),dpi=250)
plt.pie(df.Attacktype.value_counts().drop("Unknown"),labels=df.Attacktype.value_counts().drop("Unknown").index,autopct='%.2f%%',explode=[0.02 for i in range(8)])

plt.show()
```

# Favourite Target Type

```python
Targett = df['Targettype'].value_counts()
Targett['Others'] = Targett[Targett[:] <= 1000].sum()
Targett.drop(Targett[Targett[:] <= 1000].index, inplace=True)
Targett

```

```python
plt.figure(figsize=(16,22))
plt.pie(Targett.tolist(), labels=Targett.index.tolist(), autopct='%1.1f%%',shadow=True, startangle=0)
plt.title('Distribution according to Target Type\n', weight='bold')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.countplot(df['Targettype'],palette='rainbow_r',order=df['Targettype'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Favorite Targets', weight='bold')
plt.show()
```

# Terrorist Groups with Highest Terror Attacks

```python
TGroup=df['Group'].value_counts()
```

```python
plt.subplots(figsize=(15,9))
sns.barplot(TGroup[1:21].values, TGroup[1:21].index)
plt.xticks(rotation=90)
plt.title('Terrorist Groups with Highest Terror Attacks')
plt.show()
```

```python
df['Group'].value_counts().drop('Unknown').head(10).plot(kind='bar',color='green',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.show()
```

# number of kills by each terrorist group in each country Visualization

```python

ndf=df[['Group','Country','casualities']]
ndf=ndf.groupby(['Group','Country'],axis=0).sum().sort_values('casualities',ascending=False).drop('Unknown').reset_index().head(10)
ndf
```

```python
# number of kills by each terrorist group in each country Visualization
ndf=df[['Group','Country','casualities',"kill","Wound"]]
ndf=ndf.groupby(['Group','Country'],axis=0).sum().sort_values('casualities',ascending=False).drop('Unknown').reset_index().head(10)
ndf
```

```python
plt.figure(figsize=(20,6),dpi=100)
ax=sns.barplot(x='Country',y='casualities',data=ndf,ci=False)
for i in ax.containers:
    ax.bar_label(i,fmt='%.2f')
    
plt.show()
```

# Most Affected city
